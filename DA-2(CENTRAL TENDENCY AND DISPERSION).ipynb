```python
#Quartile
a=np.array([1,2,3,4,5])
#Return 25th Percentile
q1=np.percentile(a,25)
print(q1)
#Return 50th Percentile
q2=np.percentile(a,50)
print(q2)
#Return 75th Percentile
q3=np.percentile(a,75)
print(q3)
```
